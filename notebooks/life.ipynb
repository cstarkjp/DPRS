# **Cellular automaton in Rust-Python**

Demo integration of parallelized Rust into Python. Parallelization is achieved using [`Rayon`](https://docs.rs/rayon/latest/rayon/).

The core Rust module is a simple cellular automaton model – Conway's Game of Life – adapted from code provided [by the `Rayon` developers](https://github.com/rayon-rs/rayon/tree/main/rayon-demo/src/life).

Note: this notebook can be run simply using `ipython notebooks/life.ipynb` if you first `uv pip install nbformat` along with the obvious.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from numpy.typing import NDArray
```

```python
from dprs import sim
sim
```

```python
n_x = 60
n_y = 40
n_iterations = 1_000
```

```python
raw_lattice: list = sim.life(n_x, n_y, n_iterations,)
lattice: NDArray = np.array(raw_lattice, dtype=np.bool,).reshape(n_y, n_x,).T
```

```python
color_map = ListedColormap(((0.8,0.8,0.8,), (0.7,0,0.7),))
plt.imshow(((lattice.T)), cmap=color_map, origin="lower",)
color_bar = plt.colorbar(
    ticks=(0.25, 0.75,), shrink=0.5*(n_y/n_x)**0.25, aspect=15,
)
color_bar.set_ticklabels([0,1])
plt.grid(ls=":")
plt.show()
```
