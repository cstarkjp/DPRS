# **Directed percolation in Rust-Python**

First stab at implementing DP in Rust with a Python convenience wrapper.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from numpy.typing import NDArray
from dprs import sim
```

```python
params = dict(
    n_x = 3_000,
    n_y = 5_000,
    # n_z = 1,
    p = 0.5,
    n_iterations = 100,
    serial_skip = 10,
    n_threads = 16,
)
```

Execute the Rust-coded sim:

```python
n_lattices: int
raw_lattices: list[list] 
(n_lattices, raw_lattices,)= sim.dp(**params)
```

The sim return a subset of the time-series of lattices generated by the model run, where the subsetting rate is given by the `sample_rate` parameter, and where the total number returned is `n_lattices`. Each lattice is given in flattened form as a 1d list, with all the lattices concatenated into one long 1d list.

Next, we convert this long list into a 1+2d numpy array, with the leading index being the lattice slice number.

```python
lattices: NDArray = np.array(raw_lattices, dtype=np.bool,) \
                    .reshape(n_lattices, params["n_x"], params["n_y"],)
```

Function to plot the lower-left portion of the lattice:

```python
def image_lattice(lattices: NDArray, i_lattice: int=0, x: int=61, y: int=41,):
    fig = plt.figure()
    color_map = ListedColormap(((0.8, 0.8, 0.8,), (0.7, 0, 0.7),))
    plt.imshow(
        lattices[i_lattice, 0:x,0:y].T, 
        cmap=color_map, origin="lower",
    )
    color_bar = plt.colorbar(
        ticks=(0.25, 0.75,), 
        shrink=0.5*(params["n_y"]/params["n_x"])**0.25, 
        aspect=15,
    )
    color_bar.set_ticklabels((0, 1,),)
    plt.grid(ls=":")
    # plt.close()
    return fig, plt
```

```python
fig, plt = image_lattice(lattices, 10,)
plt.show()
```
